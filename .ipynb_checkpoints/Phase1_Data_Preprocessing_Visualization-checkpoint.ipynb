# Phase 1 — Data Preprocessing & Feature Engineering
## SWE485: Selected Topics in Software Engineering
### Obesity or CVD Risk Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)

print('All libraries imported successfully.')
```

## 1. Load the Dataset

```python
df = pd.read_csv('Dataset/raw_kaggle_data.csv')

print(f'Dataset shape: {df.shape}')
print(f'\nColumn names:')
print(df.columns.tolist())
df.head()
```

## 2. Initial Data Inspection

```python
print('=== Data Types & Non-Null Counts ===')
print(df.info())

print('\n=== Missing Values Per Column ===')
print(df.isnull().sum())

print('\n=== Duplicate Rows ===')
print(f'Duplicates: {df.duplicated().sum()}')

print('\n=== Target Variable Distribution ===')
print(df['NObeyesdad'].value_counts())
```

## 3. Data Cleaning
### 3.1 Remove Duplicate Rows

```python
df.drop_duplicates(inplace=True)

print(f'Rows after removing duplicates: {len(df)}')
print(f'New shape: {df.shape}')
```

### 3.2 Drop Height and Weight After Engineering BMI

```python
# First engineer BMI from Height and Weight
# BMI = Weight(kg) / Height(m)^2
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

# drop Height and Weight
df.drop(columns=['Height', 'Weight'], inplace=True)

print('BMI engineered and Height/Weight dropped!')
print(f'New shape: {df.shape}')
print(df['BMI'].describe())
```

## 4. Feature Engineering
### 4.1 Label Encode Categorical Variables

```python
le = LabelEncoder()
categorical_cols = ['Gender', 'family_history_with_overweight',
                    'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC',
                    'MTRANS', 'NObeyesdad']
label_mappings = {}

for col in categorical_cols:
    df[col + '_encoded'] = le.fit_transform(df[col])
    label_mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))
    print(f'{col}: {label_mappings[col]}')

df.drop(columns=categorical_cols, inplace=True)
print(f'\nEncoding done! New shape: {df.shape}')
```

### 4.2 Normalize Numerical Features

```python
cols_to_normalize = ['Age', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE', 'BMI']

scaler = MinMaxScaler()
df[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])

print('Normalization done! Sample:')
df.head()
```

### 4.3 Save Preprocessed Dataset

```python
df.to_csv('Dataset/preprocessed_data.csv', index=False)

print(f'Preprocessed dataset saved!')
print(f'Final shape: {df.shape}')
print(f'\nFinal columns:')
print(df.columns.tolist())
```

 ## 5. Visualizations
### 5.1 Class Distribution Plot

```python
obesity_map = {v: k for k, v in label_mappings['NObeyesdad'].items()}
df_plot = df.copy()
df_plot['Obesity Level'] = df_plot['NObeyesdad_encoded'].map(obesity_map)

fig, ax = plt.subplots(figsize=(10, 6))
counts = df_plot['Obesity Level'].value_counts()
bars = ax.bar(counts.index, counts.values,
              color=['#4C9BE8','#E8844C','#4CE87A',
                     '#E84C4C','#9B4CE8','#E8D44C','#4CE8D4'],
              edgecolor='white', linewidth=1.2)

for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 3,
            str(int(bar.get_height())),
            ha='center', fontsize=11, fontweight='bold')

ax.set_title('Distribution of Obesity Level Classes', fontsize=14, fontweight='bold')
ax.set_xlabel('Obesity Level')
ax.set_ylabel('Count')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('Supervised_Learning/plots/class_distribution.png', dpi=150)
plt.show()
print('Insight: Classes are very well balanced — ~270-350 per class!')
```

### 5.2 Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(14, 10))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask, annot=True, fmt='.2f',
            cmap='coolwarm', linewidths=0.5, ax=ax,
            annot_kws={'size': 7})

ax.set_title('Correlation Heatmap of All Features',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('Supervised_Learning/plots/correlation_heatmap.png', dpi=150)
plt.show()
print('Insight: BMI and family history show strongest correlation with obesity level!')
```

### 5.3 Boxplots: BMI and Age by Obesity Level

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(data=df_plot, x='Obesity Level', y='BMI',
            hue='Obesity Level', legend=False, ax=axes[0])
axes[0].set_title('BMI by Obesity Level', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Obesity Level')
axes[0].set_ylabel('BMI (normalized)')
axes[0].tick_params(axis='x', rotation=30)

sns.boxplot(data=df_plot, x='Obesity Level', y='Age',
            hue='Obesity Level', legend=False, ax=axes[1])
axes[1].set_title('Age by Obesity Level', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Obesity Level')
axes[1].set_ylabel('Age (normalized)')
axes[1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.savefig('Supervised_Learning/plots/boxplots_bmi_age.png', dpi=150)
plt.show()
print('Insight: BMI clearly increases across obesity levels — a strong predictor!')
```

### 5.4 Feature Distribution Histograms

```python
features_to_plot = ['Age', 'BMI', 'FAF', 'CH2O', 'NCP', 'TUE']

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for i, feature in enumerate(features_to_plot):
    axes[i].hist(df[feature], bins=20,
                 color='#4C9BE8', edgecolor='white', linewidth=0.8)
    axes[i].set_title(f'Distribution of {feature}',
                      fontsize=11, fontweight='bold')
    axes[i].set_xlabel(f'{feature} (normalized)')
    axes[i].set_ylabel('Frequency')

plt.suptitle('Feature Distributions After Normalization',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('Supervised_Learning/plots/feature_distributions.png', dpi=150)
plt.show()
print('Insight: FAF and TUE show skewed distributions — most people have low physical activity!')
```

### 5.5 Pairplot of Key Features

```python
pairplot_features = ['BMI', 'Age', 'FAF', 'CH2O', 'Obesity Level']

pp = sns.pairplot(df_plot[pairplot_features],
                  hue='Obesity Level',
                  palette='Set1',
                  plot_kws={'alpha': 0.5},
                  diag_kind='kde')
pp.figure.suptitle('Pairplot of Key Features by Obesity Level',
                   fontsize=13, fontweight='bold', y=1.02)
plt.savefig('Supervised_Learning/plots/pairplot.png', dpi=150)
plt.show()
print('Insight: BMI shows the clearest separation between obesity classes!')
```

### 5.6 BMI Category by Obesity Level

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.boxplot(data=df_plot, x='Obesity Level', y='BMI',
            hue='Obesity Level', legend=False, ax=ax)
sns.stripplot(data=df_plot, x='Obesity Level', y='BMI',
              color='black', alpha=0.2, size=2, ax=ax)

ax.set_title('BMI Distribution by Obesity Level',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Obesity Level')
ax.set_ylabel('BMI (normalized)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('Supervised_Learning/plots/bmi_by_class.png', dpi=150)
plt.show()
print('Insight: BMI is our strongest engineered feature — validates its creation!')
```
